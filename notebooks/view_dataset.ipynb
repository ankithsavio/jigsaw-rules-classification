```python
import pandas as pd

TRAIN_DATA_PATH = "../dataset/train.csv"
TEST_DATA_PATH = "../dataset/test.csv"
```

```python
train_df = pd.read_csv(TRAIN_DATA_PATH)
train_df.head()
```

```python
def view_random_sample(df):
    sample = df.sample()
    row = None
    for key, value in sample.to_dict().items():
        # grab row index
        if not row:
            row = list(value.keys())[0]
        print(f"\n{key}")
        print(f"{value[row]}")
```

```python
view_random_sample(train_df)
```
