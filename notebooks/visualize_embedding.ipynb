```python
import json
with open("../dataset/result/embedding.json", "r") as file:
    embedding_collection = json.load(file)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
embedding_collection.keys()
```

```python
def vis(column):
    embeddings = embedding_collection[column]
    embeddings = np.array(embeddings)
    plt.style.use('dark_background')
    plt.scatter(embeddings[:,0], embeddings[:,1], s=5)
    plt.gca().set_aspect('equal', 'datalim')
    plt.title(f"{column.upper()} Embedding Distribution")
    plt.show()
```

```python
vis('body')
```
